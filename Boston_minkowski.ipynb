```python
import numpy as np
from sklearn import datasets
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
import pandas
```

```python
dataset = datasets.load_boston()
X_data = dataset.data
y_data = dataset.target
X_data, y_data = shuffle(X_data, y_data)
X_data[0]
y_data[0]
```

```python
X_data = scale(X_data)
X_data[0]
```

```python
KFold_generator = KFold(n_splits=5, shuffle=True, random_state=42)
accuracy = []
lin = np.linspace(1, 10, 200)
for p in lin:
    regr = KNeighborsRegressor(n_neighbors=5, weights='distance', metric='minkowski', p=p)
    cross_val = cross_val_score(regr, X_data, y_data, scoring='neg_mean_squared_error', cv=KFold_generator)
    print('       metric {}'.format(p), cross_val)
    acc_score = np.average(cross_val)
    accuracy.append(acc_score)

max_acc = accuracy[0]
answer_index = 0
for index, element in enumerate(accuracy):
    print('     ', index, element)
    if element > max_acc:
        max_acc = element
        answer_index = index

print()
print('       max_acc', max_acc)
print('       answer_index', answer_index)
```
