```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
import pandas
```

```python
#dataset = pandas.read_csv('wine.data.csv')
names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols',
        'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue',
         'OD280/OD315 of diluted wines', 'Proline']
dataset = pandas.read_csv('wine.data.csv', names=names)
dataset = shuffle(dataset)
dataset.head()
```

```python
y_data = dataset.loc[:, ['Class']]
X_data = dataset.loc[:, ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols',
        'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue',
         'OD280/OD315 of diluted wines', 'Proline']]
#X_data.head()
#y_data.head()
y_data = np.ravel(y_data)
#y_data
```

```python
X_data_scaled = scale(X_data)
X_data_scaled
```

```python
KFold_generator = KFold(n_splits=5, shuffle=True, random_state=42)

X_datas = [X_data, X_data_scaled]
for X in X_datas:
    #print('         ', X)
    accuracy = []
    i = 1
    while i <= 50:
        neighbors_clf = KNeighborsClassifier(n_neighbors=i)
        cross_val = cross_val_score(neighbors_clf, X, y_data, cv=KFold_generator)
        #print('       neighbors {}'.format(i), cross_val)
        acc_score = np.average(cross_val)
        accuracy.append(acc_score)
        i += 1
    
    max_acc = accuracy[0]
    answer_index = 0
    for index, element in enumerate(accuracy):
        #print('     ', index, element)
        if element > max_acc:
            max_acc = element
            answer_index = index

    print()
    print('       max_acc', max_acc)
    print('       answer_index', answer_index)
```
